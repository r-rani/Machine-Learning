```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans #for now go from scratch later
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#variables 
Ks = [2, 3, 10, 20, 40]
stopping_theta = 0.001
seed = 1131
image1= plt.imread("img1lab3.jpg")
image1 = image1.astype(np.float32) / 255.0
image2 = mpl.image.imread("img2lab3.jpeg")
image2 = image2.astype(np.float32) / 255.0
methods = ["random", "random", "spread"]
max_iters=300
```

```python
def flatten_rgb(img): #flatten to a 2d vector of shape 
    H,W,three= image.shape
    X=img.reshape(-1,3)
    return X, H,W
```

```python
def unflatten(X, H, W):
    #(N,3) -> (H,W,3)
    return np.clip(X.reshape(H, W, 3), 0.0, 1.0)
```

```python
#2 initialization functions 
def init_centers_random(X, K):
    """
    Randomly selects K distinct data points from X as initial cluster centers.

    Parameters:
        X : ndarray, shape (N, 3)
            Flattened RGB pixels.
        K : int
            Number of clusters.

    Returns:
        centers : ndarray, shape (K, 3)
            Randomly chosen cluster centers.
    """
    N = X.shape[0] #N is the size 
    indices = np.random.choice(N, K, replace=False) #randomly chooses K, non replaced/unique, indices from 0-N-1
    centers = X[indices] #chosen indices 
    return centers
```

```python
def init_centers_spread(X, K):
    """
    Farthest-first (spread) initialization for K-means centers.

    Parameters:
        X : ndarray, shape (N, 3)
            Flattened RGB pixels.
        K : int
            Number of clusters.

    Returns:
        centers : ndarray, shape (K, 3)
            Well-spread initial cluster centers.
    """
    N = X.shape[0]

    # 1. Randomly pick the first center
    first_idx = np.random.randint(0, N)
    centers = [X[first_idx]]

    # 2. Iteratively pick farthest points
    for _ in range(1, K):
        # Compute squared distances from each point to its nearest chosen center
        distances = np.min(
            np.linalg.norm(X[:, None, :] - np.array(centers)[None, :, :], axis=2),
            axis=1
        )
        # Pick the point with the maximum distance
        next_center = X[np.argmax(distances)]
        centers.append(next_center)

    return np.array(centers)
```

```python
def euclid(a, b):
    """Return Euclidean distance ||a - b|| for 1D arrays a,b of length 3."""
    return float(np.linalg.norm(a - b))
```

```python
def dist_point_to_centers(x, centers):
    """
    x: (3,)
    centers: (K,3)
    returns: list of length K with Euclidean distances
    """
    K = centers.shape[0]
    dists = [0.0] * K
    for k in range(K):
        dists[k] = euclid(x, centers[k])
    return dists
    
```

```python

def assign_clusters_euclidean(X, centers):
    """
    Assigns each data point to the nearest center using Euclidean distance.
    Slower but easy to understand.
    """
    N = X.shape[0]
    K = centers.shape[0]
    labels = np.zeros(N, dtype=int)

    for i in range(N):
        # Compute Euclidean distance from point i to all K centers
        distances = np.linalg.norm(X[i] - centers, axis=1)
        labels[i] = np.argmin(distances)
    
    return labels
```

```python
def reseed_empty_center_farthest(X, centers):
    """
    Pick the data point that is farthest from its nearest CURRENT center.
    X: (N,3), centers: (K,3)
    returns one (3,) center
    """
    N = X.shape[0]
    max_min_dist = -1.0
    farthest_idx = 0

    for i in range(N):
        # distance from X[i] to nearest center
        nearest = float("inf")
        for k in range(centers.shape[0]):
            d = euclid(X[i], centers[k])
            if d < nearest:
                nearest = d
        # track the point with largest nearest-center distance
        if nearest > max_min_dist:
            max_min_dist = nearest
            farthest_idx = i

    return X[farthest_idx].copy()
```

```python
def update_centers_euclidean(X, labels, centers):
    """
    Recompute centroids as means of assigned points; reseed empties.
    X: (N,3), labels: (N,), centers: (K,3) from previous iter
    returns: (C_new, empty_indices)
    """
    K = centers.shape[0]
    C_new = np.zeros_like(centers)
    counts = np.zeros(K, dtype=int)

    # accumulate sums per cluster
    for i in range(X.shape[0]):
        k = labels[i]
        C_new[k] += X[i]
        counts[k] += 1

    empty_idx = []
    for k in range(K):
        if counts[k] > 0:
            C_new[k] /= counts[k]
        else:
            # reseed via farthest-from-nearest-center rule (Euclidean)
            C_new[k] = reseed_empty_center_farthest(X, centers)
            empty_idx.append(k)

    return C_new, empty_idx
```

```python
def init_centers_random(X, K, seed=None):
    rng = np.random.default_rng(seed)
    idx = rng.choice(X.shape[0], K, replace=False)
    return X[idx].copy()
```

```python
def init_centers_spread(X, K, seed=None):
    rng = np.random.default_rng(seed)
    N = X.shape[0]
    centers = np.empty((K, 3), dtype=X.dtype)
    # first center random
    centers[0] = X[rng.integers(0, N)]
    chosen = 1
    # iteratively add farthest from nearest chosen center
    while chosen < K:
        farthest_idx = 0
        farthest_min_dist = -1.0
        for i in range(N):
            # distance to nearest already-chosen center
            nearest = float("inf")
            for j in range(chosen):
                d = euclid(X[i], centers[j])
                if d < nearest:
                    nearest = d
            if nearest > farthest_min_dist:
                farthest_min_dist = nearest
                farthest_idx = i
        centers[chosen] = X[farthest_idx]
        chosen += 1
    return centers
```

```python
def rmse_pixels(X, Xhat):
    """sqrt(mean(||x - xhat||^2))"""
    sse = 0.0
    for i in range(X.shape[0]):
        d = X[i] - Xhat[i]
        sse += float(np.dot(d, d))
    return (sse / X.shape[0]) ** 0.5
```

```python
def run_image(img, img_name="image"):
    X = img.reshape(-1, 3).astype(np.float32)     # flatten to (N,3)
    H, W = img.shape[:2]

    runs = []                 # all runs (for logging/trace)
    best_per_k = {}           # K -> dict with best recon among inits

    for K in Ks:
        for init in methods:
            # 1) initialize centers
            if init == "random":
                centers = init_centers_random(X, K, seed)
            else:  # "spread"
                centers = init_centers_spread(X, K, seed)

            # 2) iterate K-means
            rmse_prev  = float("inf")
            rmse_trace = []
            converged  = False

            for it in range(1, max_iters + 1):
                labels = assign_clusters_euclidean(X, centers)
                centers, empties = update_centers_euclidean(X, labels, centers)
                if len(empties) > 0:
                    print(f"[{img_name}] K={K} {init}: reseeded empty clusters {empties}")

                # reconstruct + RMSE
                Xhat = np.empty_like(X, dtype=np.float32)
                for i in range(X.shape[0]):
                    Xhat[i] = centers[labels[i]]
                rmse_curr = rmse_pixels(X, Xhat)
                rmse_trace.append(rmse_curr)

                # stop on relative improvement
                if rmse_prev < float("inf"):
                    rel = (rmse_prev - rmse_curr) / rmse_prev if rmse_prev > 0 else 0.0
                    if rel < stopping_theta:
                        converged = True
                        break
                rmse_prev = rmse_curr

            img_rec = np.clip(Xhat.reshape(H, W, 3), 0.0, 1.0)

            # store this run
            run_info = {
                "K": K, "init": init, "rmse": rmse_curr, "iters": len(rmse_trace),
                "converged": converged, "img_rec": img_rec, "rmse_trace": rmse_trace
            }
            runs.append(run_info)
            print(f"[{img_name}] K={K:>2}  init={init:<6}  iters={len(rmse_trace):>3}  "
                  f"conv={converged!s:<5}  RMSE={rmse_curr:.6f}")

            # keep best reconstruction for this K
            if (K not in best_per_k) or (rmse_curr < best_per_k[K]["rmse"]):
                best_per_k[K] = run_info

    # ---- Plot: original + best reconstruction for each K ----
    Ks_sorted = sorted(best_per_k.keys(), key=lambda k: Ks.index(k))  # keep your Ks order
    n_cols = len(Ks_sorted) + 1
    plt.figure(figsize=(3.2*n_cols, 3.2))

    # original
    plt.subplot(1, n_cols, 1)
    plt.imshow(img); plt.axis("off")
    plt.title(f"{img_name}: original")

    # one panel per K (best among inits)
    for j, K in enumerate(Ks_sorted, start=2):
        r = best_per_k[K]
        plt.subplot(1, n_cols, j)
        plt.imshow(r["img_rec"]); plt.axis("off")
        plt.title(f"K={K}\nRMSE={r['rmse']:.4f}, it={r['iters']}")

    plt.tight_layout(); plt.show()

    # (optional) also show RMSE trace of the first run
    if len(runs) > 0:
        r0 = runs[0]
        plt.figure()
        plt.plot(range(1, len(r0["rmse_trace"])+1), r0["rmse_trace"], marker='o')
        plt.xlabel("Iteration"); plt.ylabel("RMSE")
        plt.title(f"{img_name}: RMSE trace (K={r0['K']}, init={r0['init']})")
        plt.grid(True, linestyle=":")
        plt.show()

    return runs, best_per_k
```
