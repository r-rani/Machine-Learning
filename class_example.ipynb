```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
#print(diabetes.DESCR)
X_data, t = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, t_train, t_test = train_test_split(X_data, t, test_size = 0.2, random_state = 10)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
plt.scatter(X_train[:,2], t_train, color='magenta')
plt.show(0)
```

```python
N=len(X_train)
M=len(X_test)
#print (N,M)
```

```python
new_col = np.ones(N);
#print (new_col.shape)
```

```python
X = np.insert(X_train, 0, new_col, axis=1)
#print(X[:4,:4])
```

```python
A=np.dot(X.T, X)
#print(A.shape)
```

```python
A1=np.linalg.inv(A)
#print(A1)
```

```python
b=np.dot(X.T, t_train)
print (b.shape)
```

```python
w=np.dot(A1,b)
print(w)
```

```python
y = np.dot(X,w)
diff_train=np.subtract(t_train,y)
err_train=np.dot(diff_train,diff_train)/N
RMSE_train = np.sqrt(err_train)
print(RMSE_train)
```

```python
new_col2 = np.ones(M);
X2 = np.insert(X_test, 0, new_col2, axis=1)
A2=np.dot(X2.T, X2)
A12=np.linalg.inv(A2)
b2=np.dot(X2.T, t_test)
w2=np.dot(A12,b2)
y2 = np.dot(X2,w2)
diff_test=np.subtract(t_test,y2)
err_test=np.dot(diff_test,diff_test)/M
RMSE_test = np.sqrt(err_test)
print(RMSE_test)
```

```python
#Plotting for 1 feature
X_one_train = X_train[:,2]
X_one_train = np.reshape(X_one_train, (N,1))
#print(X_one_train.shape)
```

```python
X_train_plot = np.insert(X_one_train, 0, new_col, axis=1)
print(X_train_plot.shape)
```

```python
#train error 
A_train_plot=np.dot(X_train_plot.T, X_train_plot)
A1_train_plot=np.linalg.inv(A_train_plot)
b_train_plot=np.dot(X_train_plot.T, t_train)
w_train_plot=np.dot(A1_train_plot,b_train_plot)
y_train_plot = np.dot(X_train_plot,w_train_plot)
diff_train_plot=np.subtract(t_train,y_train_plot)
err_train_plot=np.dot(diff_train_plot,diff_train_plot)/N
RMSE_train_plot = np.sqrt(err_train_plot)
print(err_train_plot)
print(RMSE_train_plot)
#you can see that the error here is 53->74 and 2300->5500 so higher
```

```python
#compute the test error 
X_one_test = X_test[:,2]
X_one_test = np.reshape(X_one_test, (M,1))
X_test_plot = np.insert(X_one_test, 0, new_col2, axis=1)
print(X_test_plot.shape)
A_test_plot=np.dot(X_test_plot.T, X_test_plot)
A1_test_plot=np.linalg.inv(A_test_plot)
b_test_plot=np.dot(X_test_plot.T, t_test)
w_test_plot=np.dot(A1_test_plot,b_test_plot)
y_test_plot = np.dot(X_test_plot,w_test_plot)
diff_test_plot=np.subtract(t_test,y_test_plot)
err_test_plot=np.dot(diff_test_plot,diff_test_plot)/M
RMSE_test_plot = np.sqrt(err_test_plot)
print(err_test_plot)
print(RMSE_test_plot)
```

```python
plt.scatter(X_train[:,2], t_train, color='magenta')
plt.scatter(X_train[:,2], y_train_plot, color='green') #y_one is predicted values
plt.scatter(X_test[:,2], t_test, color='blue') #test data 
plt.show()
```
